<a id='Linien'></a>
### Linien (Typ , Farbe , Dicke)

<div class="alert alert-block alert-info">
<b>Hinweis zur Linienfarbe:</b> Wenn die Linienfarbe selbstständig definiert wird, so wird für jeden neuen plot selbstständig eine neue Liniefarbe gewählt 
</div>

**`ls` - Linienstil**
<img src="figures/ls.png" width=300px>

**`c` - Linienfarbe** 
<img src="figures/colors.png" width=300px>

**`lw` - Liniendicke**
default=1

Bilder aus: [offizielle CheatSheets von Nicolas P. Rougier](https://github.com/matplotlib/cheatsheets/)

```python
plt.plot(x,y,ls="-",c="blue",lw=1);
plt.plot(x,y2,ls="--",c="red",lw=3);
```

<a id='Marker'></a>
### Marker (Typ , Farbe , Dicke)

**`marker` - Markerstil** 
<img src="figures/markers.png" width=300px>

**`markevery` - Anzahl**
<img src="figures/markevery.png" width=300px>

**`ms` - Markergröße**

```python
plt.plot(x,y,marker="o",ms=10);
plt.plot(x,y2,marker="s",ms=15,markevery=[0,-1]);
```
